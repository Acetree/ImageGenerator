```python
## 排斥关系
exclusion = {
    "Black":["Brown Bob","Black Bob"]
}
```

```python
## Generate Traits

TOTAL_IMAGES = 11000 # Number of random unique images we want to generate

all_images = [] 

# A recursive function to generate unique image combinations
def create_new_image():
    
    new_image = {} 
    
    # For each trait category, select a random trait based on the weightings 

    # 肤色选定
    # 通过skin，确定hand候选
    
    skin_picked = random.choices(skin, skin_weights)[0]
    hand_tmp_list = skin_match_hand[skin_picked]["list"]
    
    if len(hand_tmp_list) > 0:
        hand_tmp_weight = skin_match_hand[skin_picked]["weight"]
        hand_picked = random.choices(hand_tmp_list, hand_tmp_weight)[0]
        
        #objects
        objects_tmp_list = hand_match_objects[hand_picked]["list"]  
        if len(objects_tmp_list) > 0:
            objects_tmp_weight = hand_match_objects[hand_picked]["weight"]
            objects_picked = random.choices(objects_tmp_list, objects_tmp_weight)[0]
        else:
            objects_picked = ""
    else:
        hand_picked = ""
        
    new_image ["Skin"] = skin_picked
    new_image ["Hand"] = hand_picked
    new_image ["Object"] = objects_picked
        
        
    # 头发选定
    # 通过front_hair，确定back_hair候选，back_hair可以没有
    front_hair_picked = random.choices(front_hair, front_hair_weights)[0]
    back_hair_tmp_list = front_match_back[front_hair_picked]["list"]
    
    if len(back_hair_tmp_list) > 0:
        back_hair_tmp_weight = front_match_back[front_hair_picked]["weight"]
        back_hair_picked = random.choices(back_hair_tmp_list, back_hair_tmp_weight)[0]
    else:
        back_hair_picked = ""
    
    new_image ["Front Hair"] = front_hair_picked
    new_image ["Back Hair"] = back_hair_picked
    
    

    new_image ["Background"] = random.choices(background, background_weights)[0]
    new_image ["Cloth"] = random.choices(cloth, cloth_weights)[0]
    new_image ["Glasses"] = random.choices(glasses, glasses_weights)[0]
    new_image ["Mouth"] = random.choices(mouth, mouth_weights)[0]
    
    currentList = []
    for attr in new_image:
        currentList.append(new_image[attr])
        
#     print(currentList)
    
    for attr in new_image:
        if new_image[attr] in exclusion:
            for a in exclusion[new_image[attr]]:
                if a in currentList:
#                     print(currentList)
                    return False
    
    
    if new_image in all_images:
        return False
    else:
        return new_image
    
    
# Generate the unique combinations based on trait weightings
for i in range(TOTAL_IMAGES): 
    
    new_trait_image = create_new_image()
    if new_trait_image:
        all_images.append(new_trait_image)
    

print(all_images)
```

```python
# Returns true if all images are unique
def all_images_unique(all_images):
    seen = list()
    return not any(i in seen or seen.append(i) for i in all_images)

print("Are all images unique?", all_images_unique(all_images))
```

```python
# Get Trait Counts

background_count = {}
for item in background:
    background_count[item] = 0
    
back_hair_count = {}
for item in back_hair:
    back_hair_count[item] = 0

skin_count = {}
for item in skin:
    skin_count[item] = 0
    
cloth_count = {}
for item in cloth:
    cloth_count[item] = 0
    
front_hair_count = {}
for item in front_hair:
    front_hair_count[item] = 0

glasses_count = {}
for item in glasses:
    glasses_count[item] = 0
    
mouth_count = {}
for item in mouth:
    mouth_count[item] = 0
    
object_count = {}
for item in objects:
    object_count[item] = 0
    
hand_count = {}
for item in hand:
    hand_count[item] = 0

for image in all_images:
    background_count[image["Background"]] += 1
    if image["Back Hair"] in back_hair_count:
        back_hair_count[image["Back Hair"]] += 1
    skin_count[image["Skin"]] += 1
    cloth_count[image["Cloth"]] += 1
    front_hair_count[image["Front Hair"]] += 1
    glasses_count[image["Glasses"]] += 1
    mouth_count[image["Mouth"]] += 1
    if image["Object"] in object_count:
        object_count[image["Object"]] += 1
    if image["Hand"] in hand_count:
        hand_count[image["Hand"]] += 1
    
print('Total Number:',len(all_images))
print(background_count)
print(back_hair_count)
print(skin_count)
print(cloth_count)
print(front_hair_count)
print(glasses_count)
print(mouth_count)
print(object_count)
print(hand_count)
```

```python
#### Generate Metadata for each Image    

f = open('./metadata/all-traits.json',) 
data = json.load(f)


IMAGES_BASE_URI = ".../Lady/"
PROJECT_NAME = "Lady"

def getAttribute(key, value):
    return {
        "trait_type": key,
        "value": value
    }
for i in data:
    token_id = i['tokenId']
    token = {
        "image": IMAGES_BASE_URI + str(token_id) + '.png',
        "tokenId": token_id,
        "name": PROJECT_NAME + ' ' + str(token_id),
        "attributes": []
    }
    token["attributes"].append(getAttribute("Background", i["Background"]))
    token["attributes"].append(getAttribute("Back Hair", i["Back Hair"]))
    token["attributes"].append(getAttribute("Skin", i["Skin"]))
    token["attributes"].append(getAttribute("Cloth", i["Cloth"]))
    token["attributes"].append(getAttribute("Front Hair", i["Front Hair"]))
    token["attributes"].append(getAttribute("Glasses", i["Glasses"]))

    with open('./metadata/' + str(token_id), 'w') as outfile:
        json.dump(token, outfile, indent=4)
f.close()
```
