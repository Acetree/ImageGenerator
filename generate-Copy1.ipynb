```python
# Each image is made up a series of traits
# The weightings for each trait drive the rarity and add up to 100%

background = ["Gray Layers", "Dark Circle", "Galaxy"] 
background_weights = [1, 1, 98]

back_hair = ["Straight Long"] 
# back_hair_weights = [100]

skin = ["White"] 
skin_weights = [100]

cloth = ["Spectrum","LV","Sport"] 
cloth_weights = [30, 60, 10]

front_hair = ["Spectrum","Straight Short","Straight Long"] 
front_hair_weights = [30, 60, 10]

glasses = ["Spectrum","Head Set","Sunglasses"] 
glasses_weights = [30, 60, 10]

mouth = ["Green"] 
mouth_weights = [100]

objects = ["Circle"] 
object_weights = [100]

hand = ["White"] 
hand_weights = [100]

# Dictionary variable for each trait. 
# Eech trait corresponds to its file name

background_files = {
    "Gray Layers": "bg01",
    "Dark Circle": "bg02",
    "Galaxy": "bg03",
}

back_hair_files = {
    "Straight Long": "backhair3",
}

skin_files = {
    "White": "skin01",
}

cloth_files = {
    "Spectrum": "cloth01",
    "LV": "cloth02",
    "Sport": "cloth03",
}

front_hair_files = {
    "Spectrum": "fronthair01",
    "Straight Short": "fronthair02",
    "Straight Long": "fronthair03",
}

glasses_files = {
    "Spectrum": "glasses01",
    "Head Set": "glasses02",
    "Sunglasses": "glasses03",
}

mouth_files = {
    "Green": "mouth02",
}

object_files = {
   "Circle": "object01",
}

hand_files = {
    "White": "hand01",
}

```

```python
## 依赖关系和排斥关系
back_hair_dependence = {
    # front_hair : [back_hair1,back_hair2...]
    "Spectrum": {
        "list":[],
        "weight":[],
    },
    "Straight Short": {
        "list":["","Straight Long"],
        "weight":[60,40],
    },
    "Straight Long": {
        "list":["Straight Long"],
        "weight":[100],
    },
}
```

```python
## Generate Traits

TOTAL_IMAGES = 1000 # Number of random unique images we want to generate

all_images = [] 

# A recursive function to generate unique image combinations
def create_new_image():
    
    new_image = {} 
    
    # For each trait category, select a random trait based on the weightings 

    # 肤色选定
    # 通过skin，确定hand候选
    new_image ["Skin"] = random.choices(skin, skin_weights)[0]
    new_image ["Hand"] = random.choices(hand, hand_weights)[0]

    new_image ["Background"] = random.choices(background, background_weights)[0]
    new_image ["Cloth"] = random.choices(cloth, cloth_weights)[0]
    new_image ["Glasses"] = random.choices(glasses, glasses_weights)[0]
    new_image ["Mouth"] = random.choices(mouth, mouth_weights)[0]
    new_image ["Object"] = random.choices(objects, object_weights)[0]
    
    # 头发选定
    # 通过front_hair，确定back_hair候选，back_hair可以没有
    front_hair_picked = random.choices(front_hair, front_hair_weights)[0]
    back_hair_tmp_list = back_hair_dependence[front_hair_picked]["list"]
    
    if len(back_hair_tmp_list) > 0:
        back_hair_tmp_weight = back_hair_dependence[front_hair_picked]["weight"]
        back_hair_picked = random.choices(back_hair_tmp_list, back_hair_tmp_weight)[0]
    else:
        back_hair_picked = ""
    
    new_image ["Front Hair"] = front_hair_picked
    new_image ["Back Hair"] = back_hair_picked
    
    
    
    if new_image in all_images:
        return False
    else:
        return new_image
    
    
# Generate the unique combinations based on trait weightings
for i in range(TOTAL_IMAGES): 
    
    new_trait_image = create_new_image()
    if new_trait_image:
        all_images.append(new_trait_image)
    

print(all_images)
```

```python
# Returns true if all images are unique
def all_images_unique(all_images):
    seen = list()
    return not any(i in seen or seen.append(i) for i in all_images)

print("Are all images unique?", all_images_unique(all_images))
```

```python
# Add token Id to each image
i = 0
for item in all_images:
    item["tokenId"] = i
    i = i + 1
```

```python
# Get Trait Counts

background_count = {}
for item in background:
    background_count[item] = 0
    
back_hair_count = {}
for item in back_hair:
    back_hair_count[item] = 0

skin_count = {}
for item in skin:
    skin_count[item] = 0
    
cloth_count = {}
for item in cloth:
    cloth_count[item] = 0
    
front_hair_count = {}
for item in front_hair:
    front_hair_count[item] = 0

glasses_count = {}
for item in glasses:
    glasses_count[item] = 0
    
mouth_count = {}
for item in mouth:
    mouth_count[item] = 0
    
object_count = {}
for item in objects:
    object_count[item] = 0
    
hand_count = {}
for item in hand:
    hand_count[item] = 0

for image in all_images:
    background_count[image["Background"]] += 1
    if image["Back Hair"] in back_hair_count:
        back_hair_count[image["Back Hair"]] += 1
    skin_count[image["Skin"]] += 1
    cloth_count[image["Cloth"]] += 1
    front_hair_count[image["Front Hair"]] += 1
    glasses_count[image["Glasses"]] += 1
    mouth_count[image["Mouth"]] += 1
    object_count[image["Object"]] += 1
    hand_count[image["Hand"]] += 1
    
print('Total Number:',len(all_images))
print(background_count)
print(back_hair_count)
print(skin_count)
print(cloth_count)
print(front_hair_count)
print(glasses_count)
print(mouth_count)
print(object_count)
print(hand_count)
```

```python
#### Generate Metadata for all Traits 
METADATA_FILE_NAME = './metadata/all-traits.json'; 
with open(METADATA_FILE_NAME, 'w') as outfile:
    json.dump(all_images, outfile, indent=4)
```

```python
  
    
#### Generate Images    
for item in all_images:

    blank_image = Image.new(mode="RGBA", size=(1024, 1024))
    
    if item["Background"]!= "":
        im1 = Image.open(f'./trait-layers/01bg/{background_files[item["Background"]]}.png').convert('RGBA')
    else:
        im1 = blank_image
    
    if item["Back Hair"]!= "":
        im2 = Image.open(f'./trait-layers/02backhair/{back_hair_files[item["Back Hair"]]}.png').convert('RGBA')
    else:
        im2 = blank_image
        

    if item["Skin"]!= "":
        im3 = Image.open(f'./trait-layers/03skin/{skin_files[item["Skin"]]}.png').convert('RGBA')
    else:
        im3 = blank_image
        
        
  
    
    
    im4 = Image.open(f'./trait-layers/08object/{object_files[item["Object"]]}.png').convert('RGBA')
    im5 = Image.open(f'./trait-layers/09hand/{hand_files[item["Hand"]]}.png').convert('RGBA')
    
    im6 = Image.open(f'./trait-layers/04cloth/{cloth_files[item["Cloth"]]}.png').convert('RGBA')
    im7 = Image.open(f'./trait-layers/05fronthair/{front_hair_files[item["Front Hair"]]}.png').convert('RGBA')
    im8 = Image.open(f'./trait-layers/06glasses/{glasses_files[item["Glasses"]]}.png').convert('RGBA')
    
    im9 = Image.open(f'./trait-layers/07mouth/{mouth_files[item["Mouth"]]}.png').convert('RGBA')

    #Create each composite
    com1 = Image.alpha_composite(im1, im2)
    com2 = Image.alpha_composite(com1, im3)
    com3 = Image.alpha_composite(com2, im4)
    com4 = Image.alpha_composite(com3, im5)
    com5 = Image.alpha_composite(com4, im6)
    com6 = Image.alpha_composite(com5, im7)
    com7 = Image.alpha_composite(com6, im8)
    com8 = Image.alpha_composite(com7, im9)
    

    #Convert to RGB
    rgb_im = com8.convert('RGB')
    file_name = str(item["tokenId"]) + ".png"
    rgb_im.save("./images/" + file_name)
    
    
    
```

```python
#### Generate Metadata for each Image    

f = open('./metadata/all-traits.json',) 
data = json.load(f)


IMAGES_BASE_URI = ".../Lady/"
PROJECT_NAME = "Lady"

def getAttribute(key, value):
    return {
        "trait_type": key,
        "value": value
    }
for i in data:
    token_id = i['tokenId']
    token = {
        "image": IMAGES_BASE_URI + str(token_id) + '.png',
        "tokenId": token_id,
        "name": PROJECT_NAME + ' ' + str(token_id),
        "attributes": []
    }
    token["attributes"].append(getAttribute("Background", i["Background"]))
    token["attributes"].append(getAttribute("Back Hair", i["Back Hair"]))
    token["attributes"].append(getAttribute("Skin", i["Skin"]))
    token["attributes"].append(getAttribute("Cloth", i["Cloth"]))
    token["attributes"].append(getAttribute("Front Hair", i["Front Hair"]))
    token["attributes"].append(getAttribute("Glasses", i["Glasses"]))

    with open('./metadata/' + str(token_id), 'w') as outfile:
        json.dump(token, outfile, indent=4)
f.close()
```
